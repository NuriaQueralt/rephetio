```python
# Enns et al. 
# explore edge and node types
enns_df = pd.read_table('./data/ngly1-associations/manual-curation/enns-2014-network.tsv')
enns_df = (enns_df
               [['Source', 'Predicate', 'Target', 'Source_type', 'Target_type', 'Source_name', 'Target_name']]
)
#enns_df[['Source_type', 'Target_type']].groupby(['Source_type', 'Target_type']).apply(np.unique)

edges_df = enns_df
edges_df = (edges_df
                .assign(type = 
                    lambda row: edges_df['Source_type'] + '_' + edges_df['Target_type'].str.replace(' ', '-'))
)
edge_df = pd.DataFrame({'types':list(set(edges_df.type))})
print(len(edge_df.types))
edge_df
```
