# Map diseases: orphanet to do 

## Map Ophanumbers to DO (MONDO xref)

```python
# read disease_gene associations from orphadata
orphadata ='/home/nuria/workspace/repurposing/rephetio-su/orphanet/data/orphanet-disease-gene.tsv'
orphadata_df = pd.read_table(orphadata)

# read mappings from mondo
mappings = '/home/nuria/workspace/repurposing/rephetio-su/orphanet/data/mondo-orpha2do-mappings.tsv'
mappings_df = pd.read_table(mappings)

# map diseases to do
orphadata_df['orphanet_code'] = 'Orphanet:' + orphadata_df['orphanet_code'].astype(str)
orphadata_do_df = orphadata_df.merge(mappings_df, left_on='orphanet_code', right_on='orphanet', how='left')
orphadata_do_df.head()
```

```python
orphadata_do_df.shape
```

```python
orphadata_do_mapped_df = orphadata_do_df[orphadata_do_df['do'].notnull()]
```

```python
orphadata_do_null_df = orphadata_do_df[orphadata_do_df['do'].isnull()]
```

```python
do_df = orphadata_do_mapped_df[['do', 'orphanet_term', 'orphanet_gene_id',
       'gene_orphanumber', 'gene_name', 'gene_type', 'gene_symbol', 'dga_type',
       'dga_status']]
do_df = do_df.rename(columns={'do':'orphanet_code'})
do_df.shape
```

```python
orpha_df = orphadata_do_null_df[['orphanet_code', 'orphanet_term', 'orphanet_gene_id',
       'gene_orphanumber', 'gene_name', 'gene_type', 'gene_symbol', 'dga_type',
       'dga_status']]
orpha_df.shape
```

```python
orphanet_mapped_df = pd.concat([do_df,orpha_df], ignore_index=True)
orphanet_mapped_df.shape
```
